```python
import numpy as np
from scipy.stats import qmc
from math import floor
import pandas as pd

import os

n_samples=60000
divider=n_samples/6

dim=15

base=qmc.LatinHypercube(d=dim,seed=2).random(n=n_samples)
X=base #store random samples in an array
X=np.array(X)

#Identifying parameter L and a

L_min,L_max=1.0,10.0
dL=0.05

L_grid=np.arange(L_min,L_max+(1e-10),dL)

L_idx=np.floor(X[:,0]*len(L_grid))
L_idx=np.array([int(L_elt) for L_elt in L_idx])

X[:,0]=np.round(L_grid[L_idx],2)

#Identifying parameter L and a

a_min,a_max=0.05,0.95
da=0.05

a_grid=np.arange(a_min,a_max+(1e-10),da)

a_idx=np.floor(X[:,1]*len(a_grid))
a_idx=np.array([int(a_elt) for a_elt in a_idx])

X[:,1]=np.round(a_grid[a_idx],2)

valid_id=[k for k in range(0,18)]
valid_id.extend([k for k in range(42,90)])
valid_id=np.array(valid_id)

sec_idx=np.floor(X[:,2]*len(valid_id))
sec_idx=np.array([int(sec_elt) for sec_elt in sec_idx])

X[:,2]=valid_id[sec_idx]

valid_names=['IPE80', 'IPE100', 'IPE120', 'IPE140', 'IPE160', 'IPE180', 'IPE200', 'IPE220', 'IPE240', 'IPE270', 'IPE300', 'IPE330', 'IPE360', 'IPE400', 'IPE450', 'IPE500', 'IPE550', 'IPE600',
             'HEA100', 'HEA120', 'HEA140', 'HEA160', 'HEA180', 'HEA200', 'HEA220', 'HEA240', 'HEA260', 'HEA280', 'HEA300', 'HEA320', 'HEA340', 'HEA360', 'HEA400', 'HEA450', 'HEA500', 'HEA550', 'HEA600', 'HEA650', 'HEA700', 'HEA800', 'HEA900', 'HEA1000',
             'HEB100', 'HEB120', 'HEB140', 'HEB160', 'HEB180', 'HEB200', 'HEB220', 'HEB240', 'HEB260', 'HEB280', 'HEB300', 'HEB320', 'HEB340', 'HEB360', 'HEB400', 'HEB450', 'HEB500', 'HEB550', 'HEB600', 'HEB650', 'HEB700', 'HEB800', 'HEB900', 'HEB1000']

id_list=list(X[:,2].astype(int))

name_list=[valid_names[np.where(valid_id == id_list[i])[0][0]] for i in range(len(id_list))]
family_list=[name_list[i][0:3] for i in range(len(name_list))]

X[:,3:15]=10**(-1+6*X[:,3:15])

L_case=[floor(1+i/divider) for i in range(n_samples)]

print(L_case.count(1),L_case.count(2),L_case.count(3),L_case.count(4),L_case.count(5),L_case.count(6),
      L_case.count(1)+L_case.count(2)+L_case.count(3)+L_case.count(4)+L_case.count(5)+L_case.count(6)==n_samples)

L_case=np.array(L_case).reshape(-1,1)

X=np.hstack((L_case,X))

ID=[i for i in range(n_samples)]
ID=np.array(ID).reshape(-1,1)

X=np.hstack((ID,X))

input_data=pd.DataFrame({
    "ID"        : X[:,0].astype(int),
    "Lcase"     : X[:,1].astype(int),
    "SecID"     : id_list,
    "SecFamily" : family_list,
    "SecName"   : name_list,
    "L"         : X[:,2],
    "a"         : X[:,3],
    "kappa_x_A" : X[:,5],
    "kappa_x_B" : X[:,6],
    "kappa_y_A" : X[:,7],
    "kappa_y_B" : X[:,8],
    "kappa_z_A" : X[:,9],
    "kappa_z_B" : X[:,10],
    "zeta_x_A"  : X[:,11],
    "zeta_x_B"  : X[:,12],
    "zeta_y_A"  : X[:,13],
    "zeta_y_B"  : X[:,14],
    "zeta_z_A"  : X[:,15],
    "zeta_z_B"  : X[:,16]})

input_data_script=pd.DataFrame({
    "ID"        : X[:,0].astype(int),
    "Lcase"     : X[:,1].astype(int),
    "SecID"     : id_list,
    "L"         : X[:,2],
    "a"         : X[:,3],
    "kappa_x_A" : X[:,5],
    "kappa_x_B" : X[:,6],
    "kappa_y_A" : X[:,7],
    "kappa_y_B" : X[:,8],
    "kappa_z_A" : X[:,9],
    "kappa_z_B" : X[:,10],
    "zeta_x_A"  : X[:,11],
    "zeta_x_B"  : X[:,12],
    "zeta_y_A"  : X[:,13],
    "zeta_y_B"  : X[:,14],
    "zeta_z_A"  : X[:,15],
    "zeta_z_B"  : X[:,16]})
```

```python
file_path_metadata = "/input_data_V2.csv"

try:
    # Ensure the directory exists
    os.makedirs(os.path.dirname(file_path_metadata), exist_ok=True)
    
    # Save DataFrame to CSV
    input_data.to_csv(file_path_metadata, index=False, encoding="utf-8")
    print(f"DataFrame successfully saved to '{file_path_metadata}'")
except PermissionError:
    print("Permission denied: Unable to write to the specified location.")
except FileNotFoundError:
    print("Invalid file path.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

#Adapted version for Grasshopper manipulation :

file_path_script = "/input_data_script_V2.csv"

try:
    # Ensure the directory exists
    os.makedirs(os.path.dirname(file_path_script), exist_ok=True)
    
    # Save DataFrame to CSV
    input_data_script.to_csv(file_path_script, index=False, encoding="utf-8")
    print(f"DataFrame successfully saved to '{file_path_script}'")
except PermissionError:
    print("Permission denied: Unable to write to the specified location.")
except FileNotFoundError:
    print("Invalid file path.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
